```python
import pystac
import xarray as xr
import geopandas as gpd
import pandas as pd

stac_endpoint_url = 'https://s3.waw3-1.cloudferro.com/emodnet/stac/catalog.json'

# Create an empty DataFrame to store the ARCO assets
arco_assets_list = []

catalog = pystac.Catalog.from_file(stac_endpoint_url)

# Iterate over collections in the catalog
for collection in catalog.get_children():
    try:
        # Access items in the collection
        for item in collection.get_all_items():
            for asset_key, asset in item.assets.items():
                if asset.href.endswith('.zarr') or asset.href.endswith('.zarr/') or asset.href.endswith('.parquet'):
                    # Append ARCO asset information to the list
                    arco_assets_list.append({'Item ID': item.id, 'Asset Key': asset_key, 'Asset Href': asset.href})
    except Exception as e:
        print(f'Problem with collection {collection.id}: {str(e)}')

# Convert the list of dictionaries to a DataFrame
arco_assets_df = pd.DataFrame(arco_assets_list)

# Display the DataFrame

print(arco_assets_df)
```
